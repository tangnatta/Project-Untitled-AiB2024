```python
# map with soil group data

FEATURE_SOIL_GROUP = ['grp_properties_upper_organic', 'grp_properties_upper_N', 'grp_properties_upper_P', 'grp_properties_upper_K', 'grp_properties_lower_organic', 'grp_properties_lower_N',
                      'grp_properties_lower_P', 'grp_properties_lower_K', 'grp_properties_upper_pH_upper', 'grp_properties_upper_pH_lower', 'grp_properties_lower_pH_upper',
                      'grp_properties_lower_pH_lower', 'grp_awc_min', 'grp_awc_max', 'grp_awc_avg']

soil_ids_search = []
soil_ids_search_grp_id = []

for idx, i in enumerate(soil_grp_df['soil_series_id'].to_list()):
    soil_ids_search += [j.lower() for j in i]
    soil_ids_search_grp_id += [idx+1]*len(i)


def find_soil_group_data(soil_series_id):
    # ex soil_series_id = 48
    soil_series_id = str(soil_series_id).strip()
    if soil_series_id[0].isnumeric():
        return soil_series_id[:2]
    try: 
        return soil_ids_search_grp_id[soil_ids_search.index(soil_series_id.lower())]
    except:
        return 61


def get_soil_group_data(grp_ids):
    # arr = [0 * len(FEATURE_SOIL_GROUP)]
    arr = []
    if len(grp_ids.split(',')) > 1:
        grp_ids = [find_soil_group_data(i) for i in grp_ids.split(',')]
        grp_id = grp_ids[0]
        # for i in grp_ids:
        #     soil_grp_df.loc[soil_grp_df['grp_id'] == i][FEATURE_SOIL_GROUP]
    else:
        grp_id = find_soil_group_data(grp_ids)
    # print(grp_id)
    return soil_grp_df[soil_grp_df['grp_id'] == int(grp_id)][FEATURE_SOIL_GROUP].values[0]


soil_feature = pd.DataFrame(df['soilgroup'].map(get_soil_group_data).to_list(), columns=FEATURE_SOIL_GROUP)

df_merge = pd.concat([df, soil_feature], axis=1)
df_merge
```

```python
# split using sklearn

X = df_merge[X_FEATURE].astype(float)
Y = df_merge[Y_FEATURE].astype(float)

X.fillna(0, inplace=True)
Y.fillna(0, inplace=True)

X = X.map(lambda x: 0 if x == -1 else x)
Y = Y.map(lambda x: 1 if x > 0 else 0)

from sklearn.model_selection import train_test_split

# from imblearn.over_sampling import RandomOverSampler


x_train, x_test, y_train, y_test = train_test_split(X.astype(np.float32), Y.astype(np.float32), test_size=0.2, random_state=9000)
```

```python
from sklearn.metrics import classification_report, top_k_accuracy_score

def y_top_k_comp(y):
    
    return np.where(y.to_numpy() == 1)[1]

def y_hat_top_k_comp(y_hat):
    y_hat = np.array(y_hat)
    if y_hat.shape[1] == 4:
        return y_hat
    return np.transpose(np.array(y_hat)[:, :, 1])

def multi_top_k_accuracy_score(y, y_hat, k=1):
    # this is handcrafted for k=1 only
    assert k == 1 
    y_hat = np.array(y_hat)
    if y_hat.shape[1] == 4:
        pred = np.argmax(y_hat, axis=1)
    else:
        pred = np.argmax(y_hat_top_k_comp(y_hat), axis=1)
    correct = 0
    for idx, i in enumerate(pred):
        if y.iloc[idx, i] == 1:
            correct += 1
    return correct / len(y)
```

```python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5)

logreg = LogisticRegression(solver='liblinear', class_weight='balanced')
logreg = OneVsRestClassifier(logreg)
logreg = CalibratedClassifierCV(
    estimator=logreg, cv=cv)
logreg = MultiOutputClassifier(logreg)
ls_y_hat_train = []
ls_y_train = []

logreg_model = logreg.fit(x_train, y_train)
y_hat_train = logreg_model.predict(x_train)
ls_y_hat_train.extend(y_hat_train)
ls_y_train.extend(y_train)
report_train = classification_report(y_train, y_hat_train)
y_hat_prob_train = logreg_model.predict_proba(x_train)
top_k_acc_train = multi_top_k_accuracy_score(y_train, y_hat_prob_train)
print('training set')
print("top k accuracy: ", top_k_acc_train)
print(report_train)

ls_y_hat_test = []
ls_y_test = []
y_hat_test = logreg_model.predict(x_test)
ls_y_hat_test.extend(y_hat_test)
ls_y_test.extend(y_test)
report_test = classification_report(y_test, y_hat_test)
y_hat_prob_test = logreg_model.predict_proba(x_test)
top_k_acc_test = multi_top_k_accuracy_score(y_test, y_hat_prob_test)
print('test set')
print("top k accuracy: ", top_k_acc_test)
print(report_test)
```
